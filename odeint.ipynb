```python
#誤差の計算

def func(y, x):
    dydx = math.sqrt(x)
    return dydx
 
x = np.arange(0, 180, 1)

y = odeint(func, 0, x)

for i in range(0,180,5):

  y2 = math.sin(math.radians(i))
  print("微分前（sin("+str(i)+")）：%.20f"%y2)

  print("odeint（sin’("+str(i)+")=cos("+str(i)+")）:%.20f"%y[i])

  T = math.cos(math.radians(i))
  print("真値（cos("+str(i)+")）：%.20f"%T)

  Error = math.sqrt((T-y[i])*(T-y[i]))
  print('誤差：%.20f\n'%Error)
```
